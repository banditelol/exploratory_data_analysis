# KNN (k-Nearest Neighborhood)

### 1. DATA TEST DAN TRAIN SAMA
- Ini adalah metode klasifikasi
- n_neighbors = 3. Mengacu pada 3 labelled data point. 
- fit() - fungsi train menggunakan dataset ortho (sama dengan data test)
- test() - fungsi test menggunakan dataset ortho (sama dengan data train)

```python
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:,data.columns != 'class'], data.loc[:,'class']
knn.fit(x,y)
prediction = knn.predict(x)
print('Prediction: {}'.format(prediction))
```

### 2. SPLIT DATA TRAIN AND TEST
- data fit (train) dan data test menggunakan dataset yang berbeda
- dilakukan dengan cara split dataset ke dalam 2 jenis data: train dan test
- data train: 70%
- data test : 30%

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=.39, random_state=1)
knn = KNeighborsClassifier(n_neighbors = 3)
x,y = data.loc[:, data.columns != 'class'], data.loc[:,'class']
knn.fit(x_train, y_train)
prediction = knn.predict(x_test)

print('With KNN (K=3) accuracy is: ', knn.score(x_test, y_test))
```

### 3. MODEL COMPLEXITY

cara ini digunakan untuk mendapatkan hyperparameter (K) yang 
akan digunakan dalam metode KNN sehingga mendapatkan akurasi terbaik

```python
neig = np.arange(1, 25)
train_acc = []
test_acc = []

# iterate over value of k (1-25)
for i,k in enumerate(neig):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(x_train, y_train)
    train_acc.append(knn.score(x_train, y_train))
    test_acc.append(knn.score(x_test, y_test))
```

```python
# plotting
plt.figure(figsize=[13,8])
plt.plot(neig, test_acc, label='test acc')
plt.plot(neig, train_acc, label='train acc')
plt.legend()

plt.title('val vs acc')
plt.xlabel('num of neighbors')
plt.ylabel('acc')
plt.xticks(neig)
# plt.savefig('')
plt.show()
```

```python
print("best accuracy is {} with K = {}".format(np.max(test_acc),1+test_acc.index(np.max(test_acc))))
```
